```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A futuristic city with flying cars, neon lights, and cyberpunk style"

image = pipe(prompt).images[0]

image
```

```python
image.save("generated_image_1.png")
```

```python
prompts = [
    "A peaceful village surrounded by mountains at sunrise",
    "A robot reading a book in a modern library"
]

for i, p in enumerate(prompts, start=2):
    img = pipe(p).images[0]
    img.save(f"generated_image_{i}.png")
    display(img)
```
